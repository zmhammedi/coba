```python
import json
import random
from multiprocessing import Process, Queue
import time
from coba.simulations import ValidationSimulation, OpenmlSimulation
from coba.learners    import RandomLearner, VowpalLearner,EpsilonBanditLearner
from coba.benchmarks  import Benchmark
from coba.benchmarks.formats import BenchmarkFileFmtV2    
```

# Testing FreeGrad (with the importance weight invariance) against other vw learners

```python
def plot_benchmark(simulations, learners):
    Benchmark(simulations).evaluate(learners).plot_learners()

text_file = open("../benchmarks/custom_openml_notakes.json", "r")
json_txt = text_file.read() 
text_file.close()

benchmark = BenchmarkFileFmtV2().filter(json.loads(json_txt))
nsets = len(benchmark._simulations)/3 # Change the 6 depending on the number of Shuffles
#result = {}
random.seed(180)

for i in range(0,int(nsets)):
    simulations = benchmark._simulations[3*i:3*(i+1)]
    learners  = [VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s --freegrad --restart'), 
                 VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s --freegrad'),
                 #VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s --freegrad --noweightinvariance'),
                 #VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s --freegrad --restart --noweightinvariance'),
                 VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s --coin'),
                 VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s -l 0.5'),
                 #VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s -l 0.05'),
                 #VowpalLearner('--cb_explore_adf --epsilon=0.1 --interactions ssa --interactions sa --ignore_linear s -l 0.005'),
                 EpsilonBanditLearner(0.1),
                 RandomLearner()
                ]
    p = Process(target=plot_benchmark, args=(simulations, learners,))
    p.start()
    # Stop a process that takes more than 100sec 
    p.join(180)
    if p.is_alive():
        p.terminate()
        p.join()
```
